### 有一筆學生資料如下:

```python
students_grade = [
    {"name": "Denji", "國文": 30, "英文": 10, "數學": 20},
    {"name": "Power", "國文": 60, "英文": 85, "數學": 70},
    {"name": "Aki", "國文": 95, "英文": 88, "數學": 92},
    {"name": "Pochita", "國文": 50, "英文": 65, "數學": 20},
    {"name": "Makima", "國文": 100, "英文": 98, "數學": 95},
    {"name": "Reze", "國文": 65, "英文": 40, "數學": 30},
]
```

##### 1. 印出所有學生姓名

```python
for i in students_grade:
    print(i["name"])
```

##### 2. 印出 `Pochita` 的數學成績

```python
for student in students_grade:
    if student["name"] == "Pochita":
        print("Pochita 的數學成績是", student["數學"])
```

#### 3. 計算每位學生的總分
輸出格式  
`Phochita 總分: xxx 分`

```python

for student in students_grade:
    score = student["國文"] + student["英文"] + student["數學"]
    print(student["name"], "總分", score)
```

#### 4. 計算每位學生的平均分數
輸出格式  
`Pochita 平均分數: xxx 分`

```python
for student in students_grade:
    if student["name"] == "Pochita":
        avrenge = (student["國文"] + student["英文"] + student["數學"]) / 3
        print("Pochita平均分數為:", avrenge)
```

#### 5. 找出數學最高分的學生

```python
max_score = 0
top_student = 0
for student in students_grade:
    if max_score < student["數學"]:
       max_score = student["數學"]
       top_student = student["name"]
    
print("數學最高分的是", top_student)   
```

#### 6. 計算全班國文平均

```python
grade = 0
for student in students_grade:
    grade += student["國文"]
print(grade // 6)
```

#### 7. 找出不及格的學生（任一科 < 60）

```python
for student in students_grade:
    if student["國文"] < 60 or student["數學"] < 60 or student["英文"] < 60:
        print("不及格的學生有:", student["name"])
```

#### 8. 新增「總分」到每個 dict
例如 :  
`{"name": "Pochita", "國文": 80, ..., "總分": 245}`

```python
#新增「總分」到每個 dict
for student in students_grade:
    student["總分"] = student["國文"] + student["英文"] + student["數學"]
print(students_grade)
```

#### 9. 找出三科都 > 80 的學生

```python
for student in students_grade:
    if student["國文"] > 80 and student["英文"] > 80 and student["數學"] > 80:
        print("三科都及格的學生有:", student["name"])
```

#### 10. 將資料轉成以下格式 (dict of dict)

```python
"""
{
    "Denji": {"國文": 30, "英文": 10, "數學": 20},
    .....
}
"""
new_dict = {}
for student in students_grade:
   x = student["name"]
   new_dict[x] = {
      "國文":student["國文"],
      "英文":student["英文"],
      "數學":student["數學"]
    }
print(new_dict)
```
