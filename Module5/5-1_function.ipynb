```python
# 建立 function 1 #沒有return 沒有參數
def greeting():
    print("Hello,World!")

greeting()
greeting()
```

```python
# 建立 function 2 #有return值的
def hello_text():
    return "good morning"

hello = hello_text()#要有變數接住這個函數
print(hello)
```

```python
# 建立 function 3 #帶參數
def hello_user(name):
    print(f"hello,{name}")

hello_user("Allen")

```

```python
# 建立 function 4 預設參數值
def say_hello(name, greeting = "good morning"):#預設參數值一定要放在後面
    print(f"{greeting} {name}")

say_hello("Allen", "hello")
```

```python
# 改寫 Module2 practice
totals = [5600, 3400, 9000, 4000]
def discount(total, discount=200):
    if total >= 2000:
        cnt = total // 2000
        total -= discount * cnt
    return total
for total in totals:
    result = discount(total)
    print(result)
    
```
