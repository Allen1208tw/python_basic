```python
# 初始化 dict
user_info = {"name": "Alice", "age": 30, "city": "New York"}
print(user_info)  # 輸出 {'name': 'Alice', 'age': 30, 'city': 'New York'}
```

```python
#資料較多的時候 建議的寫法
fruit_prices = {
    "apple":50,
    "banana":30,
    "orange":20
    }
print(fruit_prices)  # 輸出 {'apple': 50, 'banana': 30, 'orange': 20}
```

```python
# 印出 蘋果的價格
print(fruit_prices["apple"])  # 輸出 50
```

```python
# 修改 橘子的價格
fruit_prices["orange"] = 150
print(fruit_prices)  # 輸出 
```

```python
#新增kiwi
fruit_prices["kiwi"] = 80
print(fruit_prices)
```

```python
# 刪除 香蕉
del fruit_prices["banana"]
print(fruit_prices)  # 輸出 {'apple': 50, 'orange
#或是 dict.pop() 方法刪除指定的鍵值對
fruit_prices.pop("banana", None)  # 如果不存在則返回 None
print(fruit_prices)  # 輸出 {'apple': 50, 'orange': 150}
```

```python
# 建立一個 value 含有 dict 的 dict
user_info = {
    "name":"Allen",
    "age":25,
    "info":{
        "nickname":"AL",
        "roll":"keshi",
        "phone":["0912345678","0987654321"]
    }
}
print(user_info)
```

```python
# 取得 nickname 的value
print(user_info["info"]["nickname"])  # 輸出 AL

```

```python
# 取得所有 phone_number
print(user_info["info"]["phone"])
print(user_info["info"]["phone"][0])  # 輸出 0912345678
user_phone1 = user_info["info"]["phone"][0]
print(user_phone1)  # 輸出 0987654321
```

```python
'''
改寫 fruit_dict 每個水果需要 price, quantity/amount, color
apple 有 綠色 和 紅色
orange 有 橘色
tomato 有 紅色
'''
fruit_dict = {
    "apple": {
        "price": 10.0,
        "quantity": 5,
        "color": ["green", "red"]
    },
    "orange": {
        "price": 15.0,
        "quantity": 3,
        "color": ["orange"]
    },
    "tomato": {
        "price": 8.0,
        "quantity": 10,
        "color": ["red"]
    }
}
print(fruit_dict)

```

##### dict 的合併

```python
# dict 的合併
data = {
    "banana":{
        "price": 20,
        "quantity": 30,
        "color": ["yellow"]
    }
}
fruit_dict.update(data)
print(fruit_dict) 
```

##### 另外的 print 方式 `pprint` (pretty-print) 

```python
# 從 Module 引入 function
from pprint import pprint
pprint(fruit_dict)  # 使用 pprint 輸出更易讀的格式
print("="*80)
pprint(fruit_dict, sort_dicts=False)  # 不排序鍵值對
```

```python
'''
新增顏色
apple 有新的品種 顏色要有黃色
tomato 有綠色
'''

fruit_dict["apple"]["color"].remove("yellow")
print(fruit_dict["apple"]["color"]) 
```
