```python
fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['green', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': ['orange']  # 正確的寫法
    }
}
print(fruit_dict)
```

```python
# 使用 for 迴圈讀取 dict 資料1  -->dict.values
for value in fruit_dict.values():
    print(value)
```

```python
# 使用 for 迴圈讀取 dict 資料2 -->只回傳key
for i in fruit_dict:
    print(i)
```

```python
# 使用 for 迴圈讀取 dict 資料3 key,values
for key, value in fruit_dict.items():
    print(key, value)
```

```python
for key, value in fruit_dict.items():
    print(f"{key}有{','.join(value['color'])}")
```

```python
#取的dict的value另一種方式
print(fruit_dict)

#print(fruit_dict['kiwi'])沒有該key會報錯
print(fruit_dict.get('kiwi'))

if fruit_dict.get('kiwi') == None:
    print("沒有kiwi這個水果")
```

```python
# 資料處理
fruit_color_list = [
    {'apple': 'red'}, 
    {'grape': 'purple'}, 
    {'kiwi': 'yellow'}, 
    {'apple': 'green'}, 
    {'kiwi': 'green'}, 
    {'grape': 'green'}]
```

```python
#將apple的顏色整理成'red,green' grape的顏色整理成'purple,green',kiwi的顏色整理成'yellow,green'
result = {}
for fruit in fruit_color_list:
    for fruit, color in fruit.items():
        if fruit not in result:
            result[fruit] = [color]
        else:
            result[fruit].append(color)

print(result)
```

##### `defaultdict(<type>)`

```python
# 當訪問的鍵不存在時，可以自動創建鍵和預設值，而不是拋出KeyError。
from collections import defaultdict

# 以int型態的預設值 0 來初始化
result = defaultdict(int)
print(result['apple'])
print(result)
```

```python
from collections import defaultdict
result = defaultdict(list)
fruit_color_list = [
    {'apple': 'red'}, 
    {'grape': 'purple'}, 
    {'kiwi': 'yellow'}, 
    {'apple': 'green'}, 
    {'kiwi': 'green'}, 
    {'grape': 'green'}]

for fruit_color in fruit_color_list:
    for fruit, color in fruit_color.items():
        result[fruit].append(color)
result = dict(result)#強制轉型成一般的dict
print(result)
```
