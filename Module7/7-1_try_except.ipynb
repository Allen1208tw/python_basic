##### try...except...

```python
# 可以連續定義數個例外處理機制
# 最符合的例外事件類別會先拋出
try:
    x = 10

    print(y)
    print(x)#上面報錯 不執行該行程式
except: #捕捉所有的錯誤
    print("發生錯誤")
```

```python
try:
    x = 10
    print(x[0])
    print(y)
    print(x)
except NameError: #捕捉特定錯誤類別 
    print("發生錯誤")
except TypeError:
    print("TypeError")
```

```python
try:
    x = 10
    print(x[0])
    print(y)
    print(x)
    # as-->把exception產生的資料賦值給e
except Exception as e: #捕捉大部分錯誤類別
    print(f"發生錯誤:{e}") 
```

```python
try:
    x = 10
    print(x[0])
    print(y)
    print(x)
# except Exception as e: #捕捉大部分錯誤類別
    # print("發生錯誤")# 錯誤的用法 應該排在最後面才不會直接阻擋所有錯誤
except NameError: #捕捉特定錯誤類別 
    print("發生錯誤")
except TypeError:
    print("TypeError")
except Exception as e: #一律放在最下方
     print("發生錯誤")
```

```python
# 多層例外處理:
try:
    x = 10
    try: #不影響原本程式，單獨使用
        print(y)
    except NameError as e:
        print(e)
    print(x)
except Exception as e:
    print(e)
```
