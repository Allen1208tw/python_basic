```python
# 類別名稱
class car:
# 類別屬性 (Class Attribute)
    wheel = 4

    def __init__(self, brand, model, price):
        #建立實例屬性
        self.brand = brand
        self.model = model
        self.price = price
    def display_info(self):
        print(f"{self.brand}-{self.model} has {self.wheel} wheels")
        
    def show_discount_price(self,discount):
        print(f"{self.brand}-{self.model} ${int(discount*self.price)}")
# 方法 (Method)
print(car.wheel)#

car1 = car("Toyota", "camry", 700000) #建立car 
car2 = car("benz", "c300", 2700000)

print(car1.brand)
print(car1.model)
print(car1.wheel)

print(car2.brand)
print(car2.model)
print(car2.wheel)

car1.display_info()
car2.display_info()
car1.show_discount_price(0.8)
car2.show_discount_price(1.1)
```
