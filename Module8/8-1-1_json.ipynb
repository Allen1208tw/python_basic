### `json` 套件

| 用法 | 說明 |
|:--:|:--:|
|json.dump(<list/dict>, index = num ) |物件轉JSON並寫入文件|
|json.dumps(<list/dict>, indent = num) |物件轉JSON字串|
|'json.load(file)'|讀取JSON文件|
|json.loads(str) |將JSON字串轉成Python物件|

```python
import json

fruit_dict = {
    "apple": {
        "price": 10.0,
        "quantity": 5,
        "color": ["green", "red"]
    },
    "orange": {
        "price": 15.0,
        "quantity": 3,
        "color": ["orange"]
    },
    "tomato": {
        "price": 8.0,
        "quantity": 10,
        "color": ["red"]
    }
}

with open('file.json', 'w', encoding='utf-8') as f:
    f.write(json.dumps(fruit_dict, indent = 4))
```

```python
with open('file.json', 'w', encoding='utf-8') as f:
    json.dump(fruit_dict, f, indent=4)
```

```python
with open('file.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
print(data)
print(data['apple'])
```

```python
with open('file.json', 'r', encoding='utf-8') as f:
    text = f.read()#先讀檔案 確認是list or dict
    data = json.loads(text)#把字串轉成該型別的資料(只能是dict/List)

print(data)
print(data['apple'])
```
