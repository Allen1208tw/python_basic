```python
# 字串變數初始化
data = "Hello,World!"
print(data)
print(data[0])#python由零起算
```

```python
data = "hell" + "o"*5 + "," + "world!"
print("="*20)
print(data)
print("="*20)
print(len(data))
```

```python
# 替換字串
'''
str.replace(<old>, <new>)
將 string 中的 str1 替換成 str2
'''
name = "alex"
print(name)
new_name = name.replace('ex', 'len')#會 return 一個新的資料 而不是改變我原本的變數 -> 所以要用一個變數去接回傳的值
print(new_name)
```

```python
# 去除兩側空格
'''

去除字串 string 左、右兩邊的空格
'''
msg = " __Hello,World__ "
result = msg.strip("_")
print(result)
```

```python
# 字串小寫變大寫
'''

將字串 string 裡的字母全部改成小寫
'''
msg = "hello"
result = msg.lower()
print(result)


'''

將字串 string 裡的字母全部改成大寫
'''
name = "Allen"
result = name.upper()
print(result)
```

```python
# 字串分割成 list --> str.split()
greeting = "hello world! hello John"

new = greeting.split("!")
print(new)
```
