### lambda 匿名函式  
用法  

function_name = lambda arguments: expression`

````python
# 建立函式
def add(x,y):
    return x+y
result = add(5,9)
print(result)
````

````python
# 基本用法
add2 = lambda x,y : x + y
result = add(3,10)
print(result)
````

````python
# 條件判斷
def check_pass(grade):
    if grade >= 60:
        return "及格"
    else:
        return "不及格"
    
print(check_pass(80))
print(check_pass(50))
````

````python
#if達成的結果> <判斷式> else <if 未達成的結果>
check_pass2 = lambda grade : "及格" if grade >= 60 else "不及格"

print(check_pass(80))
print(check_pass(50))
````

### 小試身手  
請建立一個 list ，並在該 list 內產生多個 `[1, 2, x]` 的 list (x 的值要遞增)  
x 到 10 即可  
Ex :  
```Python
[ [1, 2, 1], [1, 2, 2], [1, 2, 3]...[1, 2, 10] ]
```

````python
list = []
for i in range(1,11):
   one_list = [1, 2, i] #每一次迴圈都在產生新的記憶體位置
   list.append(one_list)

print(list)

````

````python
list = []
for i in range(1,11):
   list.append([1, 2, i])
print(list)
````

````python
#用 list comprehension
result = []
result = [[1, 2, i] for i in range(1, 11)]
print(result)
````

##### `sort()` 方法

````python
result = [[1, 2, i] for i in range(1, 11)]
print(result)

result.sort(reverse=True)#用reverse= true改成倒數排列
print(result)
````

````python
num = [ 100, 200, 2, 5, 7, 55, 49]

num.sort()
print(num)
````

每三個 list 中間的值要改成 5  
Ex:  
```python
[ [1, 2, 1], [1, 2, 2], [1, 5, 3], [1, 2, 4], [1, 2, 5], [1, 5, 6]...]
```

````python
result = []

for i in range(1, 11):
    if i % 3 == 0:
        result.append([1, 5, i])
    else:
        result.append([1, 2, i])
print(result)
#lambda <arguments>:<expression>--->x: x[2]
result.sort(reverse=True, key=lambda x: x[2])
print(result)
````
