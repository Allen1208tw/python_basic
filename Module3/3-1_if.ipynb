條件語句 if-elif-else

```python
'''
條件判斷
if ...
if ... else ...
if ... elif ... else ...

'''
```

##### if

##### if...else...

```python
score = 90

if score == 100:
    print("滿分")
else:
    print("普普通通")
```

##### if...elif... (else)

```python
score = 90

if score == 100:
    print("滿分")
elif score >=90:
    print("優秀")
elif score >= 60:
    print("及格")
else:
    print("不及格")
```
