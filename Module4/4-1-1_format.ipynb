```python
# 多組文字
word1 = "hello"
word2 = "world"

MSG ='%s, %s!' % (word1, word2)
print(MSG)
```

```python
# 整數
price = 50
no = '蘋果 %d 元' % (price)#不能放字串會報錯
print(no)
```

```python
# 文字與整數
name = "Allen"
age = 25

result = '%s is %d years old' % (name, age)
print(result)
```

```python
# 指定寬度 (維持 10 個字元長度，預設向右看齊)
new = "[%10s]" % ("hello")
print(new)
```

```python
# 靠左對齊 (維持 10 個字元長度，向左對齊)
msg = "|| %-10s ||" % ("Hello")
print("="*13)
print(msg)
print("="*13)
```

```python
# 指定浮點數位數
result = "%.2f" % (3.14256)
print(result)
```

```python
# 指定文字長度上限 (只有文字，才在格式化字串中加入「.」來限定字串長度)
msg = "hello world"
print(f"[msg] {msg}!!")
```

```python
data = "hello"
print(f"[{data:.2}]")
```

```python
# 空白補 0
data = "hello"

print(f"[{data:_<10}]")
print(f"[{data:x>10}]")

```

```python
user_id = input("請輸入身分證號")
# user_id2 = f"{user_id:.3}"
print(f"[{user_id:*<10.3}]")
```

```python
name = "allen"
age = 18
result = f"name:{name},age: {age}"
print(result)
```

```python
name = "allen"
age = 18
result = f"name: {name:<10},age: {age:>10}"
print(result)
```
