|基本型態|創建方式|
|:---:|:---:|
| int | num = 10 |
| float | float_num = 12.5 |
| boolean | status = True/False |
| string | x = "你好" |

|容器型態|創建方式|
|:---:|:---:|
|list| x = [] |
|tuple|x = ()|

整數 (int)、浮點數 (float)

```python
num = 20
float_num = 12.5
print(num)
print(float_num)
```

布林值 (boolean)

```python
parse_ok = True
prase_not_ok = False
print(parse_ok)
print(prase_not_ok)
print(type[prase_not_ok])
```

字串 (string)

```python
str = "HELlO" 
print(str)
```

列表 (list)

```python
num_list = [1, 2, 3]
print(num_list[2])
```

```python
num_turple = (1, 2, 3)
print(num_turple)
```

元組 (tuple)

```python
num_turple = (1, 2, 3)
print(num_turple)
```

集合 (set)

```python
num_set = {1, 2, 3, 3}
print(num_set) #不會有重複的值
```

字典 (dict)

```python
user_info = {"name": "Allen", "age" : 25}
print(user_info)
print(type(user_info))
```

```python
# 創建一個空的 dict
name = {}
print(type(name))
```

```python
# 創建一個空的 set
empty_set = set()
print(empty_set)
print(type(empty_set))
```
