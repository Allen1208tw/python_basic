##### break

```python
'''
當偵測到字母 t 時，就會強制結束迴圈
'''
for char in "cantent":
    if char == "t":
        print("遇到字母t結束整個迴圈")
        break
    print(char)
```

##### continue

```python
'''
當偵測到字母 t 時，
會跳過本次迴圈剩下的程式碼 print(char)，
但不會結束迴圈，仍然會進入下一圈繼續執行
'''
for char in "cantent":
    if char == "t":
        print("遇到字母t結束整個迴圈")
        continue 
    print(char)
```

### 小試身手
計算 1 到 100 之間的和，如果值超過500就停止執行

```python
i = 0
for h in range(1,100):
    i += h
    if i >=500:
     break
print(i)
```

計算 1 到 100 之間的奇數之和，但要跳過所有被 5 整除的數

```python
i = 0
for h in range(1,101,2):
    if h % 5 == 0:
        continue
    i += h
print(i)
```
