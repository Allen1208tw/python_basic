##### 讀取檔案 read( )

```python
file = open("../Module2/2-1.py", mode='r', encoding="UTF-8")

text = file.read()

print(text)

file.close()#關閉該檔案
```

```python
# 印出 50 個 "="
print("="*50)

# 讀取檔案內容 - 常用的寫法 (執行完畢會自動關閉檔案)
#只要一離開with語句 就關閉該檔案
with open("../Module2/2-1.py", mode='r', encoding="UTF-8") as f:
    text = f.read()
print(text)
# 印出 50 個 "="
print("="*50)

```

```python
# 一行一行讀取 把每一行放進list
with open("../Module2/2-1.py", mode='r', encoding="UTF-8") as f:
    text = f.readlines()#會回傳List
print(text)
```

```python
# 資料處理
x = []
with open("../Module2/2-1.py", mode='r', encoding="UTF-8") as f:

    text = f.readlines()#會回傳List
for i in text:
    x.append(i.strip("\n"))

print(x)
```

##### 寫入檔案 write( )

```python
# 文字內容
text = '''
傑出的人會努力規劃自己的人生，普通的人則是隨便命運怎麼安排   
一個是充實、有目的的過每一天，一個是隨興的過一天算一天
創業這條路 Michael E.Gerber
'''

# 寫入檔案
with open('file.txt', 'w', encoding='utf-8') as f:
    f.write(text)


```

```python
# 寫入資料到最後一行 (附加文字在檔案內容最後一行)

text = "People say that success just happens to you. It doesn't!"
with open('file.txt', 'w', encoding='utf-8') as f:
    f.write(text+"\n")


```

```python
with open('hello.txt', mode='r', encoding='utf-8') as f:
    text = f.read()
x = text.split(" ")
y= {}
for i in x:
    if i  in y: 
        y[i] += 1
    else:
        y[i] = 1



print(y)
```
